```python
result1["n_word"]=-1
result1["n_tech"]=-1
result1["n_digital"]=-1
i=0
while i < 2814:
    text=dt.loc[i]
    stext=nltk.word_tokenize(text)
    result1.loc[i,"n_word"]=len(stext)
    wordcounts_lower = Counter(i.lower() for i in stext)
    result1.loc[i,"n_tech"]=wordcounts_lower['technology']
    result1.loc[i,"n_digital"]=wordcounts_lower['digital']
    i=i+1
```

```python
i=0
result1["n_data"]=-1
result1["n_upgrade"]=-1
result1["n_fantastic"]=-1
result1["n_great"]=-1
result1["n_definitely"]=-1
result1["n_no"]=-1
result1["n_not"]=-1
result1["n_never"]=-1
result1["n_wait"]=-1
result1["n_still"]=-1
result1["n_I"]=-1
result1["n_me"]=-1
result1["n_never"]=-1
result1["n_wait"]=-1
result1["n_still"]=-1
while i < 2814:
    text=dt.loc[i]
    stext=nltk.word_tokenize(text)
    wordcounts_lower = Counter(i.lower() for i in stext)
    result1.loc[i,"n_data"]=wordcounts_lower['data']
    result1.loc[i,"n_upgrade"]=wordcounts_lower['upgrade']
    result1.loc[i,"n_fantastic"]=wordcounts_lower['fantastic']
    result1.loc[i,"n_great"]=wordcounts_lower['great']
    result1.loc[i,"n_definitely"]=wordcounts_lower['definitely']
    result1.loc[i,"n_no"]=wordcounts_lower['no']
    result1.loc[i,"n_not"]=wordcounts_lower['not']
    result1.loc[i,"n_never"]=wordcounts_lower['never']
    result1.loc[i,"n_wait"]=wordcounts_lower['wait']
    result1.loc[i,"n_still"]=wordcounts_lower['still']
    result1.loc[i,"n_no"]=wordcounts_lower['no']
    result1.loc[i,"n_not"]=wordcounts_lower['not']
    result1.loc[i,"n_never"]=wordcounts_lower['never']
    result1.loc[i,"n_wait"]=wordcounts_lower['wait']
    result1.loc[i,"n_still"]=wordcounts_lower['still']
    i=i+1
```

```python
result1["combinetech"]=result1["n_tech"]+result1["n_digital"]+result1["n_data"]+result1["n_upgrade"]
result1["percomb"]=result1["combinetech"]/result1["n_word"]
```

```python
result1["combineposi"]=result1["n_great"]+result1["n_fantastic"]+result1["n_definitely"]
result1["percombposi"]=result1["combineposi"]/result1["n_word"]
```

```python
result1["combinenega"]=result1["n_not"]+result1["n_never"]+result1["n_no"]
result1["percombnega"]=result1["combinenega"]/result1["n_word"]
```
